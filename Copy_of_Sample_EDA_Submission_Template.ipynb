#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
numeric = hotel_booking.select_dtypes(include=['int64', 'float64']).columns
correlation = hotel_booking[numeric].corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation,fmt='.2f',annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Heat Map',fontsize=15,fontweight='bold')
plt.show()
````

##### 1. Why did you pick the specific chart?

Correlation Heatmap Purpose:

The correlation heatmap is selected as it visually illustrates the strength and direction of relationships between numeric variables in a dataset.
Visual Representation:

This tool provides a visual representation of the correlation, making it easier to interpret the connections between different features in the dataset.
Understanding Feature Relationships:

The heatmap is valuable for understanding how various features correlate with each other.
Insights into Patterns or Dependencies:

Analyzing correlations can offer insights into potential patterns or dependencies within the dataset.
Practical Utility:

It serves as a practical and efficient tool for data exploration, aiding in the identification of key relationships among numeric variables.

##### 2. What is/are the insight(s) found from the chart?

High positive correlation values (close to 1) indicate a strong positive relationship.
High negative correlation values (close to -1) indicate a strong negative relationship.
Values close to 0 suggest a weak or no correlatio
